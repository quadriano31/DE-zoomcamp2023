```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
engine.connect()
```

```python
query = """
    SELECT 1;

"""
    
```

```python
import pandas as pd
```

```python
pd.read_sql(query,engine)
```

```python
df = pd.read_csv('yellow_tripdata_2021-01.csv.gz',nrows=100)
```

```python
df.tpep_pickup_datetime = pd.to_datetime(df.tpep_pickup_datetime)
```

```python
df.tpep_dropoff_datetime = pd.to_datetime(df.tpep_dropoff_datetime)
```

```python
df.to_sql(name='yellow_tripdata_trip',con=engine,index=False,if_exists='replace')
```

```python
query = """
SELECT * FROM yellow_tripdata_trip
"""
```

```python
pd.read_sql(query,con=engine)
```

```python
green = pd.read_csv("green_tripdata_2019-01.csv.gz")
```

```python
green.head()
```

```python
green.lpep_pickup_datetime = pd.to_datetime(green.lpep_pickup_datetime)
green.lpep_dropoff_datetime = pd.to_datetime(green.lpep_dropoff_datetime)
```

```python
green.to_sql(name='green_tripdata_trip',con=engine,index=False,if_exists='replace')
```

```python
query = """
SELECT COUNT(*) FROM green_tripdata_trip
"""
pd.read_sql(query,con=engine)
```
